```python
# import requests for get
import requests
```

```python
# get data from ESPN add players page
# API URL identified via network tab from Chrome developer tools (F12)

# http://fantasy.espn.com/apis/v3/games/flb/seasons/2019/segments/0/leagues/133735?scoringPeriodId=0&view=kona_player_info
r = requests.get("http://fantasy.espn.com/apis/v3/games/flb/seasons/2019/segments/0/leagues/133735?scoringPeriodId=0&view=kona_player_info")

data = r.json()
```

```python
# exploratory

for key in data['players'][0]['player']:
    print(key)
```

```python
# function to get just 2018 stat list

def stats2018(statlist):
    for ls in statlist:
        if ls['id'] == "002018":
            return ls
```

```python
# exploratory
tempstats = data['players'][0]['player']['stats']
stats2018(tempstats)
```

```python
# create dict with key for each player, value is list of stats

stat_dict = {}

for player in data['players']:
    # two lists created, one is empty for some reason
    if 'stats' in player['player']:
        name_stats = stats2018(player['player']['stats'])
        # name_stats None if player did not play in 2018
        if name_stats is not None:
            name = player['player']['fullName']
            stat_dict[name] = name_stats
```

```python
# exploratory

for player in stat_dict:
    print(player, stat_dict[player])
```
